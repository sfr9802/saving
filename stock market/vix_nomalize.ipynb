```python
import numpy as np
import pandas as pd
import yfinance as yf
import plotly.graph_objs as go
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
df_vix = yf.download(tickers='^VIX', period = '1d', start='2017-01-01', end = '2022-10-10')
def min_max_normalize(data : list):
    normalized = []
    
    for j in data :
        normalize_num = (j-min(data))/(max(data)-min(data))
        normalized.append(normalize_num)
    return normalized

def z_score_nomalize(data : list) :
    nomalized = []
    for i in data :
        nomalize_num = (i-np.mean(data))/np.std(data)
        nomalized.append(nomalize_num)
    return nomalized

plt.plot(df_vix.index, min_max_normalize(df_vix["Close"]))
plt.figure(figsize=(40,10))
plt.show()
plt.savefig('vix_nomal.png')
```
